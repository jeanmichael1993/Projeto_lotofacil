```python
# Importar o pacotes/Frameworks

# Modelagem de dados
import pandas as pd
import numpy as np

# Datas
import datetime

# Web Scraping
import requests
from bs4 import BeautifulSoup

```

```python
URL_Lotofacil = 'https://www.gigasena.com.br/lotofacil/'
Conexao = requests.get( URL_Lotofacil )


# Coletando informações da Pagina
Coleta = BeautifulSoup( Conexao.text, 'html.parser')


Tabelas = Coleta.find_all( 'li', attrs={'class': 'gs-game_ball'} )
print(f'Foram localizadas: { len(Tabelas) } tabelas')

```

```python
numeros_loop: str = []

for x in range(len(Tabelas)):
   if Tabelas[x].text.isdigit():
    Dados = {'Numeros' : Tabelas[x].text }
    numeros_loop.append(Dados)


```

```python
pd.DataFrame( numeros_loop)
```
